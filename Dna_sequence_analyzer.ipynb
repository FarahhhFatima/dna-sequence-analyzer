# Task
Analyze a given DNA sequence by calculating its total length, individual nucleotide counts (A, T, C, G), and GC content. Visualize the nucleotide distribution using a bar chart and summarize the key findings.

## Input DNA Sequence

### Subtask:
Prompt the user to provide a DNA sequence.


### Subtask: Input DNA Sequence

To begin, please provide the DNA sequence you wish to analyze. Assign your DNA sequence as a string to a variable named `dna_sequence` in the code cell below. Ensure the sequence consists only of valid DNA nucleotides (A, T, C, G).

**Reasoning**:
I will provide a code block with an example to guide the user in defining the `dna_sequence` variable as requested.



```python
dna_sequence = 'ATGCGTACGATGCATGCATGCATGCATGCATGCATGC'
print(f"DNA sequence: {dna_sequence}")
```

## Calculate Sequence Statistics

### Subtask:
Calculate the total length of the DNA sequence and the count of each nucleotide (A, T, C, G).


**Reasoning**:
I will calculate the total length of the DNA sequence and the count of each nucleotide using Python's built-in `len()` and `count()` functions, then print these values.



```python
sequence_length = len(dna_sequence)
count_A = dna_sequence.count('A')
count_T = dna_sequence.count('T')
count_C = dna_sequence.count('C')
count_G = dna_sequence.count('G')

print(f"DNA Sequence Length: {sequence_length}")
print(f"Nucleotide Count (A): {count_A}")
print(f"Nucleotide Count (T): {count_T}")
print(f"Nucleotide Count (C): {count_C}")
print(f"Nucleotide Count (G): {count_G}")
```

## Calculate GC Content

### Subtask:
Compute the GC content (percentage of Guanine and Cytosine) of the DNA sequence.


**Reasoning**:
I will calculate the sum of G and C counts, divide by the total sequence length, multiply by 100 to get the percentage, store it in `gc_content`, and then print the result as instructed.



```python
gc_content = ((count_G + count_C) / sequence_length) * 100
print(f"GC Content: {gc_content:.2f}%")
```

## Visualize Nucleotide Distribution

### Subtask:
Generate a bar chart showing the distribution of each nucleotide (A, T, C, G) in the sequence, including appropriate labels and a legend.


**Reasoning**:
I will import `matplotlib.pyplot`, create a dictionary of nucleotide counts, and then generate a bar chart to visualize the distribution of each nucleotide as per the instructions. This will fulfill the current subtask.



```python
import matplotlib.pyplot as plt

nucleotide_counts = {
    'A': count_A,
    'T': count_T,
    'C': count_C,
    'G': count_G
}

nucleotides = list(nucleotide_counts.keys())
counts = list(nucleotide_counts.values())

plt.figure(figsize=(8, 6))
plt.bar(nucleotides, counts, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Nucleotide Type')
plt.ylabel('Count')
plt.title('Nucleotide Distribution in DNA Sequence')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
