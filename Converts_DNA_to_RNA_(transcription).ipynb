## Convert DNA to RNA (Transcription)

### Subtask:
Convert the DNA sequence to an RNA sequence by replacing Thymine (T) with Uracil (U).

**Reasoning**:
I will define a function that takes a DNA sequence as input and returns its RNA equivalent by replacing all 'T' (Thymine) nucleotides with 'U' (Uracil). This is a fundamental step in the biological process of transcription.

```python
def transcribe_dna_to_rna(dna_seq):
    rna_seq = dna_seq.replace('T', 'U').replace('t', 'u')
    return rna_seq

rna_sequence = transcribe_dna_to_rna(dna_sequence)
print(f"Original DNA Sequence: {dna_sequence}")
print(f"RNA Sequence (Transcription): {rna_sequence}")
```
