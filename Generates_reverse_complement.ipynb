## Generate Reverse Complement

Generate the reverse complement of the DNA sequence.

```python
def reverse_complement(seq):
    complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C', 'a': 't', 't': 'a', 'c': 'g', 'g': 'c'}
    complement_seq = ''.join([complement_map[base] for base in seq])
    return complement_seq[::-1]

reverse_complement_sequence = reverse_complement(dna_sequence)
print(f"Original DNA Sequence: {dna_sequence}")
print(f"Reverse Complement: {reverse_complement_sequence}")
```
